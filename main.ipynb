```python
import cv2
from Modules.Face_Mesh import FaceMeshDetector
```

```python
cap = cv2.VideoCapture(0)
detector = FaceMeshDetector()

while True:
    ret, frame = cap.read()
    if not ret:
        break

    frame, faces = detector.findFaceMesh(frame)
    print(faces)
    cv2.imshow('frame', frame)
    if cv2.waitKey(1) == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
